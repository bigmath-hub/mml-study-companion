```python
import numpy as np 

def solve_linear_system():
    """
    Solves the system of linear equation from chapter 2.1:
    4x_1 + 4x_2 = 5
    2x_1 - 4x_2 = 1
    """

    # 1. Define the coefficient matrix A
    A = np.array((
        [4, 4],
        [2, -4]
    ))

    # 2. Define the dependent variable vector b
    b = np.array([5, 1])

    # 3. Solve for x using Numpy's linear algebra module
    try:
        x = np.linalg.solve(A, b)

        print("Coefficient Matrix A:")
        print(A)
        print("\nDependent Vector b:")
        print(b)
        print("\nSolutions Vector x:")
        print(x)
    
        # Verification: Ax should be equal b
        is_correct = np.allclose(np.dot(A, x), b)
        print(f"\nVerification (Ax == b): {is_correct}")
    
    except np.linalg.LinAlgError as e:
        print(f"Error solving the system: {e}")    


```

```python
solve_linear_system()
```
