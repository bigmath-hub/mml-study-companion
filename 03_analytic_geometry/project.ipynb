```python
import numpy as np

def project_vector(x, b):
    """
    Compute the projection and residual error
    """
    # Compute lambda
    scalar_lambda = (np.dot(x, b)) / (np.dot(b, b))

    # Compute the projection
    projection = scalar_lambda * b
    
    # Compute the residual error
    residual_e = x - (scalar_lambda * b)
    check_projection = np.dot(residual_e, b)

    return projection, check_projection
```

```python
# Define x anb b
x = np.array([2, 6])
b = np.array([1, 1])

projection, check_projection = project_vector(x, b)

# Print results
print(f"The projection Vector: {projection}")
print(f"Residual error: {check_projection}")
```
