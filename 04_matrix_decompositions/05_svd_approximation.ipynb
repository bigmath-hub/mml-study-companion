```python
import numpy as np
import matplotlib.pyplot as plt

def  get_rank_k_approx(U, S, Vt, k):
    # TODO: Slice U to take only first k columns
    U_sliced = U[:, :k]

    # TODO: Slice Vt to take only first k rows
    Vt_sliced = Vt[:k, :]

    # TODO: Create diagonal matrix from first k elements of S
    Sigma_sliced = S[:k]
    Sigma_k = np.diag(Sigma_sliced)

    # TODO: return reconstruction
    A_k = U_sliced @ Sigma_k @ Vt_sliced

    return A_k
    
```

```python
# Setup: Create a structured matrix (Gradient)
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
A = np.sin(X*10) + np.cos(Y*5)

# 1. Run SVD
U, S, Vt = np.linalg.svd(A, full_matrices=False)

# 2. Reconstruct with k=1, k=5, k=20
errors = []
ks = [1, 5, 20]

print(f"Original Rank: {np.linalg.matrix_rank(A)}")

for k in ks:
    # TODO: A_k = get_rank_k_approx(...)
    A_k = get_rank_k_approx(U, S, Vt, k)

    # TODO: error = np.linalg.norm(A - A_k)
    error = np.linalg.norm(A - A_k)

    # print(f"k={k}, error={error:.4f}")
    print(f"k={k}, error={error:.4f}")    
```
